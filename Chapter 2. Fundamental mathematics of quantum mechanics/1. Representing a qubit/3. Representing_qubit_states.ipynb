## Representing Qubit States using Qiskit

```python
# Importing required packages
from qiskit.circuit import QuantumCircuit
from qiskit.visualization import plot_histogram, plot_bloch_multivector
from qiskit_aer import AerSimulator
import math
import sys
import qiskit
print(f"Python version: {sys.version}")
print(f"Qiskit version: {qiskit.__version__}")
```

```python
# Specifying the quantum simulator for simulating the circuits
backend_simulator = AerSimulator()
```

### Representing | 0 > statevector

#### Step 1: Initialise a qubit

```python
# This create a quantum circuit with one qubit
qc = QuantumCircuit(1)
# Qiskit by default initialises |0> state

# Displaying the quantum circuit
qc.draw(output="mpl")
```

#### Step 2: Determine Output State

```python
# Simulator saves the statevector
qc.save_statevector()   

# Run the simulation and return the result
result = backend_simulator.run(qc,shots=1024).result()
```

```python
# Get the output state vector
output_state = result.get_statevector()

# Display the output state vector
print(output_state) 
```

### Step 3: Measure the quantum circuit

```python
# Measure and display the quantum circuit 
qc.measure_all()
qc.draw(output="mpl")
```

#### Srep 4: Visualization of result using histogram and bloch sphere

```python
# Probability of the output
counts = result.get_counts()
plot_histogram(counts)
```

```python
counts
```

```python
# Representing the qubit on a Bloch Sphere
plot_bloch_multivector(output_state)
```

### Representing | 1 > statevector

```python
# Get the output state vector
output_state = result.get_statevector()

# Display the output state vector
print(output_state) 
```
