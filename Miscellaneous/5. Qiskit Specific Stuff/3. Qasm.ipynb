## Open Qasm
### Open Quantum Assembly Language

[Open Qasm Documentation](https://qiskit.github.io/openqasm/)

```python
from qiskit import QuantumCircuit
```

```python
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0,1)
qc.measure_all()
qc.draw('mpl')
```

#### Converting code in qiskit to Quantum Assembly Language used to create quantum circuits

```python
string = qc.qasm()
print(string)
```

#### From Qasm to Qiskit using string of qasm code

```python
qc = QuantumCircuit.from_qasm_str(string)
qc.draw('mpl')
```

#### From Qasm to Qiskit using file of qasm code

```python
filename = "qc.qasm"
qc = QuantumCircuit.from_qasm_file(filename)
```

```python
print(open(filename, "r").read())
```

```python
qc.draw('mpl')
```
