## Greenberger-Horne-Zeilinger State (GHZ)
$$\ket{GHZ} = \frac{1}{\sqrt{2}}(\ket{0}^{\otimes n} + \ket{1}^{\otimes n})$$

GHZ state is an n-qubit maximally entangled quantum state

### For n = 1, Application of H gate gives the desired result
$$\ket{GHZ} = \frac{1}{\sqrt{2}}(\ket{0} + \ket{1}) = \ket{+}$$

```python
from qiskit import QuantumCircuit, BasicAer, execute
from qiskit.visualization import array_to_latex, plot_histogram
```

```python
qc = QuantumCircuit(1)
qc.h(0)
qc.draw('mpl')
```

```python
resultant_state = execute(qc, backend=BasicAer.get_backend('statevector_simulator')).result()
array_to_latex(resultant_state.get_statevector())
```

```python
plot_histogram(resultant_state.get_counts())
```

### For n = 2, GHZ state is a Bell State with positive correlation and phase
$$\ket{GHZ} = \frac{1}{\sqrt{2}}(\ket{00} + \ket{11})$$

```python
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0,1)
qc.draw('mpl')
```

```python
resultant_state = execute(qc, backend=BasicAer.get_backend('statevector_simulator')).result()
array_to_latex(resultant_state.get_statevector())
```

```python
plot_histogram(resultant_state.get_counts())
```

### n = 3, Most trivial form of GHZ state
$$\ket{GHZ} = \frac{1}{\sqrt{2}}(\ket{000} + \ket{111})$$

```python
qc = QuantumCircuit(3)
qc.h(0)
qc.cx(0,1)
qc.cx(1,2)
qc.draw('mpl')
```

```python
resultant_state = execute(qc, backend=BasicAer.get_backend('statevector_simulator')).result()
array_to_latex(resultant_state.get_statevector())
```

```python
plot_histogram(resultant_state.get_counts())
```

### For n = 4
$$\ket{GHZ} = \frac{1}{\sqrt{2}}(\ket{0000} + \ket{1111})$$

```python
qc = QuantumCircuit(4)
qc.h(0)
qc.cx(0,1)
qc.cx(1,2)
qc.cx(2,3)
qc.draw('mpl')
```

```python
resultant_state = execute(qc, backend=BasicAer.get_backend('statevector_simulator')).result()
array_to_latex(resultant_state.get_statevector())
```

```python
plot_histogram(resultant_state.get_counts())
```
